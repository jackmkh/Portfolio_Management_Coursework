### Q6) Pick any two stock amont five stocks in Q3. Calculate the optimal weight on each stock that will provide you with highest Sharpe ratio (Assume risk-free rate is 0.5%)

```python
# Selecte the return of stock 37 and 40
df_selected = df_my[['37','41']]
df_selected.head()
```

```python
def sharpe_ratio(weights, *arg):
    '''
    weights is an array with weights of each stock
    w1 is weight of stock in first column of dataFrame
    w2 is weight of stock in second column of dataFrame
    The return data of two stock needs to be passed as *args
    The function will return (-1 * sharpe ratio) for the given weights
    '''
    df_ret = arg[0]
    portfolio_ret = np.sum(df_ret * weights, axis = 1)
    mean_portfolio_ret = np.mean(portfolio_ret)
    std_portfolio = np.std(portfolio_ret)
    sharpe = (mean_portfolio_ret - 0.0005)/std_portfolio
    return -sharpe
```

```python
res = optimize.minimize(sharpe_ratio, np.array([0.5,0.5]), args=(df_selected,), constraints={'type':'eq','fun':lambda x: x[0]+x[1]-1})
```

```python
optimal_weights = res.x
print('Optimal weight for stock 37 = ', optimal_weights[0])
print('Optimal weight for stock 41 = ', optimal_weights[1])
print('Optimal Sharpe Ratio = ', -res.fun)
```
